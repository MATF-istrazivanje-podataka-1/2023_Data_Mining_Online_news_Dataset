```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from statistics import median
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('./../data/OnlineNewsPopularity.csv')
data
```

```python
data.describe()
```

```python
data.isna().any().any()
```

```python
data.drop('url', axis = 1, inplace = True)
```

```python
Y = data[' shares']
print('Average value: ' + str((sum(Y) / len(Y))))
print('Median: ' + str(median(Y)))
```

```python
X = data.drop(' shares', axis=1)
X
```

```python
Y = [1 if i > 1400 else 0 for i in Y]
```

```python
ones = Y.count(1)
zeroes = Y.count(0)
print('Number of ones: ' + str(ones))
print('Number of zeroes: ' + str(zeroes))
```

```python
sizes = [zeroes / len(Y), ones / len(Y)]
labels = ['Not popular', 'Popular']
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels = labels, autopct = '%1.1f%%', startangle = 90)

plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt, filled=True)
plt.show()
```

```python
print(f"Default decision tree has depth {dt.get_depth()} with {dt.get_n_leaves()} nodes")
```

```python
#kod preuzet sa vezbi
plt.figure(figsize=(20, 20))
pd.Series(dt.feature_importances_, index=X.columns).plot.barh()
plt.title("Feature importance")
plt.show()
```

```python
Y_pred = dt.predict(X_test)
```
