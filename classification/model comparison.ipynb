```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
%store -r X_train
%store -r X_test
%store -r Y_train
%store -r Y_test

%store -r dtc
%store -r dtc_grid
%store -r dtc_halving_grid
%store -r random_forest
%store -r knn
%store -r knn_grid
%store -r knn_grid_iqr
```

```python
models = [dtc, dtc_grid, dtc_halving_grid, random_forest, knn, knn_grid, knn_grid_iqr]
model_names = ['dtc', 'dtc_grid', 'dtc_halving_grid', 'random_forest', 'knn', 'knn_grid', 'knn_grid_iqr']
```

```python
def plot_score_comparison(score, X, Y, text):    
    plt.figure(figsize=(15,5))

    for model, model_name in zip(models, model_names):
        Y_pred = model.predict(X)
        
        score_value = score(Y, Y_pred)
        
        bar = plt.bar(model_name, score_value)
        height = bar[0].get_height()
        plt.text(bar[0].get_x() + bar[0].get_width() / 2.0, height, f'{score_value:.4f}', ha = 'center', va = 'bottom')
        plt.title(score.__name__ + f' {text}')
```

```python
plot_score_comparison(accuracy_score, X_train, Y_train, 'train')
```

```python
plot_score_comparison(accuracy_score, X_test, Y_test, 'test')
```

```python
plot_score_comparison(precision_score, X_train, Y_train, 'train')
```

```python
plot_score_comparison(precision_score, X_test, Y_test, 'test')
```

```python
plot_score_comparison(recall_score, X_train, Y_train, 'train')
```

```python
plot_score_comparison(recall_score, X_test, Y_test, 'test')
```

```python
plot_score_comparison(f1_score, X_train, Y_train, 'train')
```

```python
plot_score_comparison(f1_score, X_test, Y_test, 'test')
```
